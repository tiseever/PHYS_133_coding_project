```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate a random array
npoints = 100
low, high = 0, 1
a = np.random.uniform(low,high,npoints)

# Sort the array in descending order
b = -a
sb= np.sort(b)
sa = -sb
```

```python
# Plot what the function looks like
lda = 10
f = -lda*np.log(1-sa)
plt.figure(figsize=(8,4))
plt.plot(f)
plt.xlabel('Point number')
plt.ylabel('Value');
```

```python
# Find the average, standard deviation, and the error on the mean
Avg = np.average(f)
Std = np.std(f)
EOM = Std/np.sqrt(npoints)
print('Average = {0}, Standard Deviation = {1}, Error on the Mean = {2}'.format(Avg,Std,EOM))
```

### This average is within predictable variation of the true average, which is ~9.08.

```python
# Create a histogram
plt.figure(figsize=(6,4))
plt.hist(f,bins=20)
plt.xlabel('Point Value')
plt.ylabel('Number of Points');
```
